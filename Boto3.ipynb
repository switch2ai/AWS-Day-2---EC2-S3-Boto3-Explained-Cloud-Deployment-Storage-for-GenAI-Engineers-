##S3 :

```python
## Client Setup (building connection)
s3 = boto3.client(
    service_name='s3',
    region_name ="ap-south-1"
    )
```

```python
##List out Buckets :
bucket_list = s3.list_buckets()

for bucket in bucket_list["Buckets"]:
  print(bucket["Name"])
```

```python
## Create bucket (Never create any resource from code) :

s3.create_bucket(
    Bucket = "velocity-260601",
    CreateBucketConfiguration = {
        "LocationConstraint" : "ap-south-1"
    }
)
```

```python
## upload file :
s3.upload_file(
    Filename = "/content/AWS.ipynb",  ## local file to upload
    Bucket = "velocity-260601",        ## bucket where file will be uploaded
    Key = "my-uploaded-file.ipynb"    ## uploaded file will be stored with this name
)
```

```python
## List out objects (files) :
objects = s3.list_objects(
    Bucket = "velocity-260601"
)

for object in objects["Contents"]:
  print(object["Key"])
```

```python
## Download file from bucket :
s3.download_file(
    Bucket = "velocity-260601",
    Key = "my-uploaded-file.ipynb",
    Filename = "/content/my-downloaded-file.ipynb"
)
```

```python
## Delete Object :
s3.delete_object(
    Bucket = "velocity-260601",
    Key = "my-uploaded-file.ipynb"
)
```
