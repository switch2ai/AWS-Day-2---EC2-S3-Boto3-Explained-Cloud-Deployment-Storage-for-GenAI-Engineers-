####IAM (Identity & Access Management) :

```python
Root user
Account owner that performs tasks requiring unrestricted access.

IAM user
User within an account that performs daily tasks.
```

```python
IAM : It allows root account to control who can access which services
      what actions can be performed on that account.
```

```python
>> User : It is an entity with credentials (Access to some services)
>> Group : It is a collection of users with same permission
>> Policy : Set of rules that allows or deny access to services
```

#### EC2 (Elastic Compute Cloud) :

```python
>> It is a virtual machine (like a computer on cloud)
   in this VM we can configure CPU, RAM, OS, GPU....

## Uses :
   -- Train Model
   -- Host API or Application
```

```python
## How to create EC2 Instance :

Launch an instance
    -- Name and tags (31-Jan)   --> Give name to EC2 instace
    -- Application and OS Images (Amazon Machine Image)  --> Select OS
    -- Instance type --> select computational requiremenst (RAM/CPU/GPU/Storage..)
    -- Key pair (login)   --> Used to establish connection with EC2 from local
    -- Network settings  --> Control traffic to your instace
    -- Configure storage
```

```python
   Inbound Rule -->  EC2 --> Outbound Rule
```

```python
## Connect with EC2 Instance :
    -- Open SSH client (CMD...)
    -- Locate your key-file (velocity.pem)
    -- Run Command
       ssh -i <name-of-key.pem> username@<public-ip-address>
```
